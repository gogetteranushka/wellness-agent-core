```python
import pandas as pd

# Load recipes with ingredients
recipes = pd.read_excel('data/recipes.xlsx')

print("="*80)
print("RECIPES DATASET (WITH INGREDIENTS)")
print("="*80)

print(f"\nüìä Shape: {recipes.shape}")
print(f"   Rows: {recipes.shape[0]}")
print(f"   Columns: {recipes.shape[1]}")

print("\nüìã COLUMN NAMES:")
for i, col in enumerate(recipes.columns, 1):
    print(f"   {i}. {col}")

print("\nüîç First 10 rows:")
print(recipes.head(10))

print("\nüîç Look at ONE complete recipe:")
# Get all rows for the first recipe
first_recipe_code = recipes['recipe_code'].iloc[0] if 'recipe_code' in recipes.columns else recipes.iloc[0, 0]
print(f"\nRecipe code: {first_recipe_code}")
print(recipes[recipes.iloc[:, 0] == first_recipe_code])
```
